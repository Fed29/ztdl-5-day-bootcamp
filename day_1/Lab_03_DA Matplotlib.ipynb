# Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def create_random_data():
    N = 1000
    data1 = np.random.normal(0, 0.1, N)
    data2 = (np.random.normal(1, 0.4, N) + np.linspace(0, 1, N))
    data3 = 2 + (np.random.random(N) * np.linspace(1, 5, N))
    data4 = (np.random.normal(3, 0.2, N) + 0.3 * np.sin(np.linspace(0, 20, N)))
    
    data = np.vstack([data1, data2, data3, data4])
    data = data.transpose()
    
    cols = ['data1', 'data2', 'data3', 'data4']

    df = pd.DataFrame(data, columns=cols)
    return df
```

```python
df = create_random_data()
```

```python
df.plot(title='Line plot');
```

```python
df.plot(style='.', title='Scatter Plot');
```

```python
df.plot(kind='hist',
        bins=50,
        title='Histogram',
        alpha=0.6);
```

```python
df.plot(kind='hist',
        bins=100,
        title='Cumulative distributions',
        density=True,
        cumulative=True,
        alpha=0.4);
```

```python
df.plot(kind='box',
        title='Boxplot');
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16,12))

df.plot(ax=ax[0][0],
        title='Line plot')

df.plot(ax=ax[0][1],
        style='o',
        title='Scatter plot')

df.plot(ax=ax[1][0],
        kind='hist',
        bins=50,
        title='Histogram')

df.plot(ax=ax[1][1],
        kind='box',
        title='Boxplot');
```

```python
categories = df['data1'] > 0.01
```

```python
counts = categories.value_counts()
counts
```

```python
counts.plot(kind='pie',
            figsize=(5, 5),
            explode=[0, 0.15],
            labels=['<= 0.1', '> 0.1'],
            autopct='%1.1f%%',
            shadow=True,
            startangle=90,
            fontsize=16);
```
